```
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns 
df=pd.read_csv("https://raw.githubusercontent.com/architsharmaa/Boston-Housing-Analysis/master/Boston.csv")
df.head()
```

```
# Density Plot
sns.kdeplot(df.ptratio, df.nox)
```

```
# Joint Distribution Plot
sns.jointplot(x='ptratio',y='nox',data=df)
```

```
#factor plotor catplot
g= sns.catplot(x='ptratio',y='nox',data=df,hue='rad',col='rad')

```

```
#swarmplot
sns.swarmplot(x='nox',y='age',data=df,hue='rad')
```

```
# Distribution Plot (a.k.a. Histogram)
sns.distplot(df.indus)
```

```
# Count Plot (a.k.a. Bar Plot)
sns.countplot(x='indus', data=df)
```

```
sns.lmplot(x='nox',y='age',data=df,fit_reg = False, hue ='rad')
```

```

sns.lmplot(x='nox',y='age',data=df,fit_reg = False, hue ='rad')
plt.ylim(0, None)
plt.xlim(0, None)
```

```
#boxplot
sns.boxplot(data=df)
```

```
#set theme
sns.set_style('whitegrid')
#violin plot
sns.violinplot(x='black',y='age',data=df)
```
